```
import plotly.express as px
```

```
import pandas as pd
```

```
df = pd.read_excel('/content/drive/MyDrive/vzlet/16vzlet.xlsx')
df.head()
```

```
fig = px.bar(df, x = 'Приложение',
             y = 'Время',
             title = 'Приложения для 16 лет' ,
             template = 'ggplot2',
             color = 'Приложение',
             hover_name = 'Приложение')
fig.update_layout(yaxis_title = 'Время',
                  xaxis_title = '',
                  title = {'font' : dict(size = 35)},
                  showlegend = False)
#fig.update_traces(hovertemplate = 'Приложение : %{x} <br> Время,проведенное в нём : %{y}')
fig.show()
```
